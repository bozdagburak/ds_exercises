```python
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
```

```python
# exercises-1
norm_dist = stats.norm(0, 1)
empty_list = []
for i in np.linspace(0,3,31):
    for j in np.arange(0,.1,.01):
        empty_list.append("%.4f" % (1-norm_dist.cdf(i+j)))
```

```python
z_table = pd.DataFrame(np.array(empty_list).reshape(len(np.linspace(0,3,31)), 
                       len(np.linspace(0,.09,10))),  index = np.linspace(0,3,31),
                       columns = np.linspace(0,.09,10))
#z_table.index.name = "k"
#z_table.columns.name = "α"
z_table
```
